```python
ax = sns.countplot(x="Gender",data =df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
purch = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
purch
```

```python
purch = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
gen = sns.barplot(x="Gender",y="Amount",data=purch)
sns.set(rc={"figure.figsize":(5,5)})
for bar in gen.containers:
    gen.bar_label(bar)
```

#### From Above graph we can see that most of buyer are female and even the purchasing power of female is greater than male.

```python
ag = sns.countplot(data = df , x="Age Group",hue="Gender")
for bar in ag.containers:
    ag.bar_label(bar)
    
```

```python
purch_ag = df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
purch_ag
```

```python
purch_ag = df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(data = purch_ag,x="Age Group",y="Amount")
```

#### From Above graph we can see that most of buyer are of age groupe between 26-35 yrs female.

```python
state_sales = df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False)
```

```python
state_sales
```

```python
state_sales = df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(15,5)})
statesales = sns.barplot(data=state_sales,x="State",y="Orders")
for bar in statesales.containers:
    statesales.bar_label(bar)
```

```python
state_sales = df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(15,5)})
stateamt = sns.barplot(data=state_sales,x="State",y="Amount")
for bar in stateamt.containers:
    stateamt.bar_label(bar)
```

#### From Above graph we can see that most of the Order and Amount are from Uttar Pradesh , Maharashtra and Karnataka respectively.

```python
occ=sns.countplot(data=df,x="Occupation")
sns.set(rc={"figure.figsize":(35,5)})
for bar in occ.containers:
    occ.bar_label(bar)
```

```python
sales_occ=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_occ=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(19,5)})
oc = sns.barplot(data=sales_occ,x="Occupation",y="Amount")
for bar in oc.containers:
    oc.bar_label(bar)
```

```python
prod=sns.countplot(data=df,x="Product_Category")
sns.set(rc={"figure.figsize":(25,5)})
for bar in prod.containers:
    prod.bar_label(bar)
```

```python
sales_prod=df.groupby(["Product_Category"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(data=sales_prod,x="Product_Category",y="Orders")
```

```python
sales_prod=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(data=sales_prod,x="Product_Category",y="Amount")
```

#### From Above graph we can see that most of sold product are from Food , Clothing and Electronics Category.

```python
sales_prod=df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
cv = sns.barplot(data=sales_prod,x="Product_ID",y="Orders")
for bar in cv.containers:
    cv.bar_label(bar)
```
